```python
import pandas as pd
import re
import xlsxwriter
```

```python
df = pd.read_excel(r"C:\Users\caio.santos\Desktop\Doc scrapping.xlsm")
```

```python
def extnum_from_DF(df):
    rex1 = re.compile('R\$\s?\d+?\,?\d?\d?\d?\.?\d?\d?\sm?b?illion')
    rex2 = re.compile('\d+?\.?\d?\d?\%')
    dic = []

    for t in df['Texto']:
        lis = []    
        if rex1.search(t):
        
            for num in rex1.findall(t):
                lis.append(num)
        if rex2.search(t):
            for num in rex2.findall(t):
                lis.append(num)
    
        dic.append(lis)       

    return df.join(pd.DataFrame(dic))

```

```python
df2 = extnum_from_DF(df)
```

```python
df2.to_excel(r"C:\Users\caio.santos\Desktop\test.xlsx")
```
